```python
input_file = csv.DictReader(open("products_bought.csv"))

global_list = []
all_user_global_list = []

count = 1

for row in input_file:
    
    print("####",count,"####")
    print(row['user_id'])
    print(row['item_category_list'])
    
    local_list = []
    all_user_local_list = []
    
    category = (dftrain['item_category_list'] == int(row['item_category_list']))
    user = (dftrain['user_id'] == int(row['user_id']))
    
    cat1 = user & category
    
    local_list.append(int(row['user_id']))
    
    for hours in [1, 2, 4, 8, 16, 32, 64, 128, 256]:
        
        high_limit = datetime.datetime.utcfromtimestamp(int(row['context_timestamp']))
        delta = datetime.timedelta(minutes = hours * 60)
        low_limit = high_limit - delta
        
        dataset = dftrain['context_timestamp']
        
        time_gap = (dataset >= low_limit) & (dataset <= high_limit)
        
        length = len(dftrain[time_gap & cat1]['user_id'])
        local_list.append(length)
        
        length = len(dftrain[time_gap & category]['user_id'])
        all_user_local_list.append(length)
        
    global_list.append(local_list)
    all_user_global_list.append(all_user_local_list)
    count = count + 1
```

```python
category_peeping = pd.DataFrame(global_list)
category_peeping.columns = ['User ID', '1', '2', '4', '8', '16', '32', '64', '128', '256']

category_peeping_all_users = pd.DataFrame(all_user_global_list)
category_peeping_all_users.columns = ['1', '2', '4', '8', '16', '32', '64', '128', '256']

category_peeping.to_csv("category_clicked.csv", index=False, encoding='utf-8')
category_peeping_all_users.to_csv("category_clicked_all_users.csv", index=False, encoding='utf-8')
```
